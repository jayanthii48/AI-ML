Finding Missing values

```python
import pandas as pd
```

```python
sale_data = pd.read_csv("Online Sales Data.csv")
```

```python
sale_data
```

```python
sale_data.isnull()
```

```python
sale_data.notnull()
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Unit Price'] = sale_data['Unit Price'].fillna(0)
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Total Revenue'].mean() 
```

```python
sale_data['Total Revenue'] = sale_data['Total Revenue'].fillna(sale_data['Total Revenue'].mean())
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Total Revenue'] 
```

```python
sale_data['Region'].mode()
```

```python
sale_data['Region']= sale_data['Region'].fillna(sale_data['Region'].mode()[0])
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Region']
```

```python
sale_data['Units Sold'].median()
```

```python
sale_data['Units Sold']=sale_data['Units Sold'].fillna(sale_data['Units Sold'].median())
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Payment Method'] = sale_data['Payment Method'].bfill()
```

```python
sale_data.isnull().sum()
```

```python
sale_data['Payment Method'] 
```

```python
sale_data.info()
```

```python
sale_data['Product category'].value_counts(dropna=False)
```

```python
sale_data[sale_data.isnull().any(axis=1)]
```

```python
sale_data[sale_data.isnull().all(axis=1)]
```

```python
sale_data.isnull().sum()
```
