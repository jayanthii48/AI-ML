```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# Sample dataset
data = {
    'Credit_Score': ['High', 'Medium', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High', 'Medium', 'Low'],
    'Income': ['High', 'Medium', 'Low', 'Medium', 'High', 'Low', 'High', 'Medium', 'Low', 'Medium'],
    'Existing_Loans': ['No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No'],
    'Employment_Status': ['Employed', 'Unemployed', 'Employed', 'Employed', 'Unemployed', 'Employed', 'Unemployed', 'Employed', 'Employed', 'Unemployed'],
    'Loan_Approved': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'No']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Encoding categorical variables
encoder = LabelEncoder()
for column in df.columns:
    df[column] = encoder.fit_transform(df[column])

# Splitting features and target
X = df.drop(columns=['Loan_Approved'])
y = df['Loan_Approved']

# Training Decision Tree model
decision_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
decision_tree.fit(X, y)

# Display the decision tree structure
print(export_text(decision_tree, feature_names=list(X.columns)))
```
