```python
import pandas as pd

df = pd.DataFrame({
    'purchase_date': pd.to_datetime([
        '2023-01-15 14:23:00',
        '2023-04-05 09:15:00',
        '2023-07-20 19:45:00'
    ])
})

df['year'] = df['purchase_date'].dt.year
df['month'] = df['purchase_date'].dt.month
df['day'] = df['purchase_date'].dt.day
df['weekday'] = df['purchase_date'].dt.weekday
df['hour'] = df['purchase_date'].dt.hour

df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)

# lambda x:  for x in range(5)

# 1 if x >= 5  else 0    if x>=5
#                           1
#                       else
#                           0
#  0 mon ,1 tue......4 fri 5 sat 6 sun


df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)

df['delivered_date'] = df['purchase_date'] + pd.to_timedelta([3, 5, 2], unit='d')
df['delivery_time_days'] = (df['delivered_date'] - df['purchase_date']).dt.days

print(df)
```

```python
df = pd.DataFrame({
    'price': [100, 200, 150],
    'tax': [5, 10, 7],
    'revenue': [1000, 2500, 1800],
    'qty': [10, 25, 18]
})

df['total_price'] = df['price'] + df['tax']
df['unit_price'] = df['revenue'] / df['qty']
df['revenue_per_price'] = df['revenue'] / df['price']

print(df)
```

```python
df = pd.DataFrame({
    'user_id': [1, 1, 2, 2, 3],
    'purchase_amt': [100, 150, 200, 300, 250]
})

# Mean purchase per user
df['user_avg_purchase'] = df.groupby('user_id')['purchase_amt'].transform('mean')

# Count of purchases per user
df['user_purchase_count'] = df.groupby('user_id')['purchase_amt'].transform('count')

print(df)
```

```python
df = pd.DataFrame({
    'purchase_amt': [120, 600, 450],
    'salary': [3000, 2500, 7000],
    'loan_emi': [1000, 900, 2000],
    'bmi': [25, 32, 29]
})

# Business logic
df['is_high_value'] = (df['purchase_amt'] > 500).astype(int)
df['salary_gt_2x_emi'] = (df['salary'] > 2 * df['loan_emi']).astype(int)
df['is_obese'] = (df['bmi'] > 30).astype(int)

print(df)
```

```python
from textblob import TextBlob

df = pd.DataFrame({
    'review': [
        "Great product, free delivery!",
        "Terrible quality. Do not buy!",
        "Okay for the price, nothing special"
    ]
})

# Word count
df['word_count'] = df['review'].apply(lambda x: len(x.split()))

#lambda x:
#len() 25
#x.split()

# Check for keyword
df['has_free'] = df['review'].str.contains('free', case=False).astype(int)

# Sentiment polarity (-1 to 1)
df['sentiment'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)

print(df)
```

```python
import numpy as np

data = [10, 12, 14, 15, 16, 17, 18, 20,-1]  # 100 is an outlier
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]
print(outliers)  # Output: [100]
print(q1 , q3)
```
